```python
# Read in an image
image = cv2.imread("Apollo_11_Launch.jpg", cv2.IMREAD_COLOR)

# Display the original image
plt.imshow(image[:,:,::-1])
```

line

```python
imageLine = image.copy()

cv2.line(imageLine, (200, 400), (400, 400), (0, 255, 0), thickness=5, lineType = cv2.LINE_AA)
plt.imshow(imageLine[:, :, ::-1])
```

circle

```python
imageCir = image.copy()

cv2.circle(imageCir, (200, 400), 100, (0, 255, 255), thickness = 4)
plt.imshow(imageCir[:, :, ::-1])
```

rectangle

```python
imageRect = image.copy()

cv2.rectangle(imageRect, (200, 400), (400, 600), (0, 255, 255), thickness = 4)
plt.imshow(imageRect[:, :, ::-1])
```

text

```python
imageText = image.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imageText, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA);

# Display the image
plt.imshow(imageText[:,:,::-1])
```
