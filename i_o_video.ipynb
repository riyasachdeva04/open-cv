```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
source = 'race_car.mp4'

cap = cv2.VideoCapture(source)
```

```python
if cap.isOpened() == False:
  print('Error streaming video')
```

```python
#display one frame
ret, frame = cap.read()
plt.imshow(frame[...,::-1])
```

```python
#display video
from IPython.display import HTML
HTML("""<video width=1024 controls>
  <source src="race_car.mp4" type="video/mp4">
</video>""")
```

```python
# Default resolutions of the frame are obtained.
# Convert the resolutions from float to integer.
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Define the codec and create VideoWriter object.
out_avi = cv2.VideoWriter('race_car_out.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))

out_mp4 = cv2.VideoWriter('race_car_out.mp4',cv2.VideoWriter_fourcc(*'XVID'), 10, (frame_width,frame_height))
```

```python
#read until complete

while(cap.isOpened()):
  # Capture frame-by-frame
  ret, frame = cap.read()
    
  if ret == True:
    
    # Write the frame to the output files
    out_avi.write(frame)
    out_mp4.write(frame)
    
  # Break the loop
  else: 
    break
```

```python
cap.release()
out_avi.release()
out_mp4.release()
```
